```python
import requests
```

```python
api_key = "d7a2805862251eb60d77d0f9e0f79922"
api_url = "https://api.themoviedb.org/3"
params = {'query':"game",'api_key': api_key}
header = {'Accept': 'applcation/json'}
```

```python
import requests
api_key = "d7a2805862251eb60d77d0f9e0f79922"
api_url = "https://api.themoviedb.org/3"
params = {'query': "christian",'api_key': api_key}
header = {'Accept': 'application/json'}
response = requests.get( api_url + "/search/person",headers = header, params = params)
```

```python
print(response.url)
```

```python
print(response.text)
```

```python
data = response.json()
result = data.get('results')
for i in result:
    if (i['name'] == 'Christian Bale'):
        print(str(i['id']) + " : " + i['name'])
```

```python
import requests
api_key = "d7a2805862251eb60d77d0f9e0f79922"
api_url = "https://api.themoviedb.org/3"
params = {'with_cast':"3894 ",'api_key': api_key,'sort_by':'popularity.desc'}
header = {'Accept': 'applcation/json'}
```

```python
response = requests.get( api_url + "/discover/movie",headers = header, params = params)
print(response.url)
print(response.text)
```

```python
data = response.json()
result = data.get('results')
for i in result:
    print(i['title'] + " : " + str(i['popularity']))
```
